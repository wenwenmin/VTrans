```python
# 正负样本划分
for k in range(1, len(m)):
    for j in range(0, len(m)-k):
        if m[j] > m[j+1]:
            m[j], m[j+1] = m[j+1], m[j]
            s[j], s[j+1] = s[j+1], s[j]
            i[j], i[j+1] = i[j+1], i[j]
m, s, i
```

```python
living_long_counts_1 = 999
living_short_counts_0 = 0
threshold = m[1]
Threshold = 999
Label_list = []
res = abs(living_long_counts_1 - living_short_counts_0)
k = 1
while threshold != max(m) and abs(living_long_counts_1 - living_short_counts_0) != 0:
    print(living_long_counts_1, living_short_counts_0, Threshold)
    Label_list_tmp = []
    #print(Label_list_tmp)
    for j in range(len(m)):
        if m[j] >= threshold:
            Label_list_tmp.append(1)
        else:
            if s[j] == '1:DECEASED':
                Label_list_tmp.append(0)
            else:
                Label_list_tmp.append(3)
    #print(Label_list_tmp)
    result = pd.value_counts(Label_list_tmp)
    if len(np.array(result).tolist()) != 3:
        k += 1
        threshold = m[k]
        continue
    living_long_counts_1 = result[1]
    living_short_counts_0 = result[0]
    if abs(living_long_counts_1 - living_short_counts_0) < res:
        Threshold = threshold
        res = abs(living_long_counts_1 - living_short_counts_0)
        Label_list = Label_list_tmp
    k += 1
    threshold = m[k]
    #print(living_long_counts_1, living_short_counts_0, threshold)
result = pd.value_counts(Label_list)
#print(living_long_counts_1, living_short_counts_0, threshold)
print(result[1], result[0], Threshold)
#living_long_counts_1, living_short_counts_0, threshold,result,Label_list
```
